### 4.2 Найдите выручку компании в зависимости от месяца
  Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, <br>а на выходе словарь, где ключи - это месяцы, а значения - это выручка.
###### _Используйте аннотирование типов._

```python

print(dot.resample('M').sum())
```

```python
import pprint 
def exit_of_month(data, summa):
    n = {}
    for i in range(len(data)):
        n[data[i]] = n.get(data[i], 0) + summa[i]
    return n
 

pprint.pprint(exit_of_month(dat, summ))
```
